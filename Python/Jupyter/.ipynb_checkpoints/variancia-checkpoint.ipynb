# Estudo de variância amostral, amplitude e desvio padrão

```python
import pandas as pd
import math

a = [15, 25, 30, 50]
a_series = pd.Series(a)
media_a = a_series.mean()
b = [0, 20, 35, 40]
b_series = pd.Series(b)
media_b = b_series.mean()

print('A: ', a)
print('média: ', media_a)
print()
print('B: ', b)
print('média: ', media_b)
```

```python
print('amplitude A: ', (a[-1] - a[0]))
print('amplitude B: ', (b[-1] - b[0]))
```

```python
# Desvio padrão de A
sum_a = 0
for i in a:
    sum_a = sum_a + math.pow((i - media_a), 2)

desvio_a = math.sqrt(sum_a / (len(a) - 1))
print('desvio-padrão A: ', int(desvio_a))

# Desvio padrão de B
sum_b = 0
for i in b:
    sum_b = sum_b + math.pow((i - media_b), 2)

desvio_b = math.sqrt(sum_b / (len(b) - 1))
print('desvio-padrão B: ', int(desvio_b))
```

```python
print('variancia de A: ', (sum_a / len(a)))
print('variancia de B: ', (sum_b / len(b)))
```

```python
a_series.plot(kind='density')
```
